```python
def figure_gen_save(train_data_df_half,file,st = 0, ed = 1000):
    i = 0
    a4_dims = (24, 20)
    for item in num_list:
        
        fig, ax = plt.subplots(1,1,figsize=a4_dims)
        ax.set_title(item,color='black',size=40)
        df_select = train_data_df_half[[item,'target']].sort_values([item])[st:ed]
        #program = "df_select = train_data_df_half[[item,'target']].sort_values(item)["+str(range)+"]"
        #exec(program)
        df_select.reset_index(inplace = True, drop = True)
        
        df_apply = df_select[df_select['target']==1]
        ax.scatter(df_apply.index,df_apply[item],c="r",marker='|',linewidths=0.5)

        df_apply = df_select[df_select['target']==0]
        #ax.scatter(df_apply.index,df_apply[item],c="blue",marker='_',linewidths=0.5)
        ax.plot(df_apply.index,df_apply[item],c="blue",linestyle='--')#,linewidths=0.5)
        ax.tick_params(axis='both', which='major', labelsize=40)
        extent = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
        fig.savefig('../huge data source/AE/figure/'+str(file)+'/' + item + '.png',facecolor='white', edgecolor='none')#,bbox_inches=extent.expanded(2, 2)
        print(str(i)+':'+ item,end="\r", flush=True)
        i = i + 1
        plt.close(fig)

def figure_gen_save_v2(train_data_df_half,file,st = 0, ed = 1000):
    i = 0
    a4_dims = (24, 20)
    for item in num_list:
        
        fig, ax = plt.subplots(1,1,figsize=a4_dims)
        ax.set_title(item,color='black',size=40)
        df_select = train_data_df_half[[item,'target']].sort_values(['target',item,])[st:ed]
        #program = "df_select = train_data_df_half[[item,'target']].sort_values(item)["+str(range)+"]"
        #exec(program)
        df_select.reset_index(inplace = True, drop = True)
        
        df_apply = df_select[df_select['target']==1]
        ax.scatter(df_apply.index,df_apply[item],c="r",marker='$+$',linewidths=0.5)

        df_apply = df_select[df_select['target']==0]
        ax.scatter(df_apply.index,df_apply[item],c="blue",marker='$+$',linewidths=0.5)
        #ax.plot(df_apply.index,df_apply[item],c="blue",linestyle='--')#,linewidths=0.5)
        ax.tick_params(axis='both', which='major', labelsize=40)
        extent = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
        fig.savefig('../huge data source/AE/figure/'+str(file)+'/' + item + '.png',facecolor='white', edgecolor='none')#,bbox_inches=extent.expanded(2, 2)
        print(str(i)+':'+ item,end="\r", flush=True)
        i = i + 1
        plt.close(fig)

def figure_GenSave_tu_item(train_data_df_half,stat,file,st = 0, ed = 1000):
    i = 0
    a4_dims = (24, 20)
    for item in num_list:
        item = (item,stat)
        fig, ax = plt.subplots(1,1,figsize=a4_dims)
        ax.set_title(item,color='black',size=40)
        df_select = train_data_df_half[[item,'target']].sort_values(['target',item,])[st:ed]

        df_select.reset_index(inplace = True, drop = True)
        
        df_apply = df_select[df_select['target']==1]
        ax.scatter(df_apply.index,df_apply[item],c="r",marker='$+$',linewidths=0.5)

        df_apply = df_select[df_select['target']==0]
        ax.scatter(df_apply.index,df_apply[item],c="blue",marker='$+$',linewidths=0.5)
        #ax.plot(df_apply.index,df_apply[item],c="blue",linestyle='--')#,linewidths=0.5)
        ax.tick_params(axis='both', which='major', labelsize=40)
        extent = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
        fig.savefig('../huge data source/AE/figure/'+str(file)+'/' + str(item) + '.png',facecolor='white', edgecolor='none')#,bbox_inches=extent.expanded(2, 2)
        print(str(i)+':'+ str(item),end="\r", flush=True)
        i = i + 1
        plt.close(fig)

def percentile(n):
    def percentile_(x):
        return np.percentile(x, n)
    percentile_.__name__ = 'percentile_%s' % n
    return percentile_

def figure_gen_save_withDatetime(train_data_df_half,file,st = 0, ed = 1000):
    i = 0
    a4_dims = (24, 20)
    pencent_list = dict({'percentile_10': 'dotted','percentile_25':'--','percentile_50':'-','percentile_75':'--','percentile_90': 'dotted'})
    
    for item in num_list:
        df_select = train_data_df_half[[item,'target','S_2_dt']]
        df_select = df_select.sort_values(item)[st:ed]
        df_select = df_select.groupby(['target','S_2_dt'],as_index=False).agg([percentile(10),percentile(25),percentile(50), percentile(75), percentile(90)]).reset_index()
        df_select = df_select.sort_values(['target','S_2_dt'])
        fig, ax = plt.subplots(1,1,figsize=a4_dims)
        ax.set_title(item,color='black',size=40)
        
        
        df_apply = df_select[df_select['target']==1].reset_index()
        
        for pen_th in pencent_list.keys():
            ax.plot(df_apply['S_2_dt'],df_apply[item][pen_th],c='r',linestyle  = pencent_list[pen_th])
            ax.scatter(df_apply['S_2_dt'],df_apply[item][pen_th],c="r",marker='x',linewidths=0.5)
        df_apply = df_select[df_select['target']==0].reset_index()
        for pen_th in pencent_list.keys():
            ax.plot(df_apply['S_2_dt'],df_apply[item][pen_th],c='blue',linestyle  = pencent_list[pen_th])
            ax.scatter(df_apply['S_2_dt'],df_apply[item][pen_th],c="blue",marker='x',linewidths=0.5)
        ax.tick_params(axis='both', which='major', labelsize=40)
        ax.tick_params(axis='x',labelrotation=90)
        #plt.show()
        extent = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
        fig.savefig('../huge data source/AE/figure/'+str(file)+'/' + item + '.png',facecolor='white', edgecolor='none')#,bbox_inches=extent.expanded(2, 2)
        
        i = i + 1
        plt.close(fig)
        print(str(i)+':'+ str(item),end="\r", flush=True)

def collective_GenSave_tu_item(df_gp,df_full,stat,file,st = 0, ed = 1000):
    i = 0
    a4_dims = (24, 20)
    for item in num_list:
        item_name = item
        df_full_select = df_full[['customer_ID',item_name,'target','period_th','period_reverth']]
        item = (item,stat) #item convert to turple....

        fig, ax = plt.subplots(1,1,figsize=a4_dims)
        ax.set_title(item,color='black',size=40)

        df_select = df_gp[[item,'target']].sort_values(['target',item,])[st:ed]
        df_select.reset_index(inplace = True)#, drop = True)

        
        #scatter the detail
        df_select = df_select.rename_axis('order').reset_index()
        df_full_select= df_full_select.join(df_select[['customer_ID','order']].set_index('customer_ID'), on='customer_ID',how='inner')
        ax.scatter(df_full_select['order'],df_full_select[item_name],c="black",marker='$.$',linewidths=0.01)
        #df_full_filter = df_full_select[df_full_select['period_th'] == 1]
        #ax.scatter(df_full_filter['order'],df_full_filter[item_name],c="green",marker='$L$',linewidths=0.2)
        df_full_filter = df_full_select[df_full_select['period_reverth'] == 1]
        ax.scatter(df_full_filter['order'],df_full_filter[item_name],c="purple",marker='$1$',linewidths=0.02)

        
        df_apply = df_select[df_select['target']==1]
        ax.scatter(df_apply.index,df_apply[item],c="r",marker='$+$',linewidths=0.5)
        df_apply = df_select[df_select['target']==0]
        ax.scatter(df_apply.index,df_apply[item],c="blue",marker='$+$',linewidths=0.5)
        
        ax.tick_params(axis='both', which='major', labelsize=40)
        extent = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
        fig.savefig('../huge data source/AE/figure/'+str(file)+'/' + str(item) + '.png',facecolor='white', edgecolor='none')#,bbox_inches=extent.expanded(2, 2)
        print(str(i)+':'+ str(item),end="\r", flush=True)
        i = i + 1
        plt.close(fig)

def q75(x):
    return x.quantile(0.75)

def q50(x):
    return x.quantile(0.50)

def q25(x):
    return x.quantile(0.25)
```
